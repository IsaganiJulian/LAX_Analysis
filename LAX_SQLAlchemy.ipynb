```python
from sqlalchemy import create_engine

# Replace 'your_username', 'your_password', 'your_host', and 'your_database' with your PostgreSQL credentials
db_uri = "postgresql://postgres:$Yahooo14@localhost/LAX"
engine = create_engine(db_uri)
```

```python
from sqlalchemy import Table, Column, Integer, String, MetaData

metadata = MetaData()

cargo_table = Table(
    'cargo_data',
    metadata,
    Column('id', Integer, primary_key=True),
    Column('Extract_Date', String),
    Column('Report_Period', String),
    Column('Arrival_Departure', String),
    Column('Domestic_International', String),
    Column('Cargo_Type', String),
    Column('Tons', Integer),
)

# Define table for terminal_df
terminal_table = Table(
    'terminal_data',
    metadata,
    Column('id', Integer, primary_key=True),
    Column('Extract_Date', String),
    Column('Report_Period', String),
    Column('Terminal', String),
    Column('Arrival_Departure', String),
    Column('Domestic_International', String),
    Column('Passenger_Count', Integer),
)

# Define table for flight_ops_df
flight_ops_table = Table(
    'flight_ops_data',
    metadata,
    Column('id', Integer, primary_key=True),
    Column('Extract_Date', String),
    Column('Report_Period', String),
    Column('Flight_Type', String),
    Column('Arrival_Departure', String),
    Column('Domestic_International', String),
    Column('Flight_Ops_Count', Integer),
)

# Create tables in PostgreSQL
metadata.create_all(engine)
```

```python
import pandas as pd
from sqlalchemy import insert

# Read cargo_df into pandas DataFrame
cargo_df = pd.read_csv(r'/Users/isaganijulian/Documents/GitHub/LAX_Analysis/cargo/cargo_df.csv')

# Convert DataFrame to a list of dictionaries
data_to_insert = cargo_df.to_dict(orient='records')

# Insert data into the PostgreSQL table
with engine.connect() as connection:
    connection.execute(insert(cargo_table).values(data_to_insert))

# Read flight_ops_df into a pandas DataFrame 
flight_ops_df = pd.read_csv(r'/Users/isaganijulian/Documents/GitHub/LAX_Analysis/flight_ops/flight_ops_df.csv')

# Convert DataFrame to a list of dictionaries
data_to_insert = flight_ops_df.to_dict(orient='records')

# Insert data into the PostgreSQL table
with engine.connect() as connection:
    connection.execute(insert(flight_ops_table).values(data_to_insert))

# Read terminal_df into a pandas DataFrame 
terminal_df = pd.read_csv(r'/Users/isaganijulian/Documents/GitHub/LAX_Analysis/terminal/terminal_df.csv')

# Convert DataFrame to a list of dictionaries
data_to_insert = terminal_df.to_dict(orient='records')

# Insert data into the PostgreSQL table
with engine.connect() as connection:
    connection.execute(insert(terminal_table).values(data_to_insert))
```
