## LAX Terminal Data Set 

```python
terminal_df = pd.read_csv(r'/Users/isaganijulian/Documents/GitHub/LAX_Project/terminal/LAX_Terminal.csv')
```

```python
# Define dictionary to map old to new column names
column_name_mapping = { 
'DataExtractDate' : 'Extract_Date', 
'ReportPeriod' : 'Report_Period', 
}

#Rename
terminal_df.rename(columns=column_name_mapping, inplace=True)

print(terminal_df)
```

```python
import pandas as pd 
from datetime import datetime 

# Convert the Extract_Date original format date 
terminal_df['Extract_Date'] = pd.to_datetime(terminal_df['Extract_Date'], format='%m/%d/%Y %I:%M:%S %p')

# Convert to desired format
terminal_df['Extract_Date'] = terminal_df['Extract_Date'].dt.strftime('%Y, %B %d, %A %I:%M:%S %p')

# Convert for Report_Period
terminal_df['Report_Period'] = pd.to_datetime(terminal_df['Report_Period'], format='%m/%d/%Y %I:%M:%S %p')

# Convert to desired format
terminal_df['Report_Period'] = terminal_df['Report_Period'].dt.strftime('%Y, %B %d, %A %I:%M:%S %p')
```

```python
# Display basic summary statistics
summary_stats = terminal_df.describe()

# Display the number of unique values in each column
unique_counts = terminal_df.nunique()

# Display the data types of each column
data_types = terminal_df.dtypes

# Display missing values count in each column
missing_values = terminal_df.isnull().sum()

# Display summary statistics
print("Summary Statistics:")
print(summary_stats)

# Display unique value counts
print("\nUnique Value Counts:")
print(unique_counts)

# Display data types
print("\nData Types:")
print(data_types)

# Display missing values
print("\nMissing Values:")
print(missing_values)
```
