```python
# Explore unique values in each categorical column
for column in data1.select_dtypes(include='object').columns:
    print(f"\nUnique values in {column}:")
    print(data1[column].value_counts())

# Plot histograms for numerical columns
data1['Passenger_Count'].plot(kind='hist', bins=20, edgecolor='black', figsize=(8, 5))
plt.title('Distribution of Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Frequency')
plt.show()

# Plot count plots for categorical columns
categorical_columns = data1.select_dtypes(include='object').columns
for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=data1, x=column)
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Pair plot for numerical columns (scatter plots and histograms)
sns.pairplot(data1, hue='Domestic_International', diag_kind='hist', height=2, aspect=2)
plt.suptitle('Pair Plot of Numerical Columns', y=2)
plt.show()

# Box plot for Passenger_Count based on Arrival_Departure
plt.figure(figsize=(10, 6))
sns.boxplot(data=data1, x='Arrival_Departure', y='Passenger_Count')
plt.title('Box Plot of Passenger Count by Arrival/Departure')
plt.xlabel('Arrival/Departure')
plt.ylabel('Passenger Count')
plt.show()
```

```python
# Reset the index temporarily to access 'Report_Period'
data1_reset = data1.reset_index()

# Convert 'Report_Period' to datetime and extract the year
data1_reset['Report_Period'] = pd.to_datetime(data1_reset['Report_Period'])
data1_reset['Year'] = data1_reset['Report_Period'].dt.year



#Loop over unique years and create individual plots
unique_years = sorted(data1_reset['Year'].unique())

for year in unique_years:
    subset_data = data1_reset[data1_reset['Year'] == year]

    plt.figure(figsize=(14, 7))
    sns.barplot(data=subset_data, x='Terminal', y='Passenger_Count', palette='viridis')
    plt.title(f'Passenger Count by Terminal for Year {year}')
    plt.xlabel('Terminal')
    plt.ylabel('Passenger Count')
    plt.xticks(rotation=45)
    plt.show()
```

 The terminals passenger count was plotted using different graphs organized by year, starting from 2006 to 2023. From the different bar plots, Terminal 1 seems to have the most passanger count based on the graphs depicted above except for the years 2007,2019, and 2021.

```python
import pandas as pd 
from tabulate import tabulate 

# Reset the index temporarily to access 'Report_Period'
data_reset = data1.reset_index()

data1_reset['Report_Period'] = pd.to_datetime(data1_reset['Report_Period'])
data1_reset['Year'] = data1_reset['Report_Period'].dt.year

# Group by Year and Terminal and calculate the sum of Passanger_Count 
grouped_data = data1_reset.groupby(['Year','Terminal'])['Passenger_Count'].sum().reset_index()

#Create a table using tabulate
table = tabulate(grouped_data, headers='keys', tablefmt='pretty', showindex=False)

print(table) 
```

This table depicts the sum of the passenger counts for the terminals based by year. 

```python
# Reset the index temporarily to access 'Report_Period'
data1_reset = data1.reset_index()

# Convert 'Report_Period' to datetime and extract the month
data1_reset['Report_Period'] = pd.to_datetime(data1_reset['Report_Period'])
data1_reset['Month'] = data1_reset['Report_Period'].dt.month_name()

# Get unique months and sort them
unique_months = sorted(data1_reset['Month'].unique())

# Loop over sorted unique months and create individual plots
for month in unique_months:
    subset_data = data1_reset[data1_reset['Month'] == month]
    
    plt.figure(figsize=(14, 7))
    sns.barplot(data=subset_data, x='Terminal', y='Passenger_Count', palette='viridis')
    plt.title(f'Passenger Count by Terminal for Month {month}')
    plt.xlabel('Terminal')
    plt.ylabel('Passenger Count')
    plt.xticks(rotation=45)
    plt.show()
```

The bar plot depicts the terminals passenger counts, ordered by the month for the years 2006-2023. 

```python
data1_reset = data1.reset_index()

# Convert 'Report_Period' to datetime and extract the month
data1_reset['Report_Period'] = pd.to_datetime(data1_reset['Report_Period'])
data1_reset['Month'] = data1_reset['Report_Period'].dt.month_name()

# Group by Month and Terminal and calculate the sum of Passenger_Count
grouped_data = data1_reset.groupby(['Month', 'Terminal'])['Passenger_Count'].sum().reset_index()

# Define the order of months
month_name_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Sort the table by month
grouped_data['Month'] = pd.Categorical(grouped_data['Month'], month_name_order, ordered=True)
grouped_data = grouped_data.sort_values(by=['Month', 'Terminal'])

# Create a table using tabulate
table = tabulate(grouped_data, headers='keys', tablefmt='pretty', showindex=False)

# Print the table
print(table)
```

This table depicts the sum of the passenger counts for the terminals ordered by months.
