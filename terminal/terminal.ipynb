## LAX Terminal Data Set 

```python
terminal_df = pd.read_csv(r'/Users/isaganijulian/Documents/GitHub/LAX_Project/terminal/LAX_Terminal.csv')
```

```python
# Define dictionary to map old to new column names
column_name_mapping = { 
'DataExtractDate' : 'Extract_Date', 
'ReportPeriod' : 'Report_Period', 
}

#Rename
terminal_df.rename(columns=column_name_mapping, inplace=True)

print(terminal_df)
```

```python
import pandas as pd 
from datetime import datetime 

# Convert the Extract_Date original format date 
terminal_df['Extract_Date'] = pd.to_datetime(terminal_df['Extract_Date'], format='%m/%d/%Y %I:%M:%S %p')

# Convert to desired format
terminal_df['Extract_Date'] = terminal_df['Extract_Date'].dt.strftime('%Y, %B %d, %A %I:%M:%S %p')

# Convert for Report_Period
terminal_df['Report_Period'] = pd.to_datetime(terminal_df['Report_Period'], format='%m/%d/%Y %I:%M:%S %p')

# Convert to desired format
terminal_df['Report_Period'] = terminal_df['Report_Period'].dt.strftime('%Y, %B %d, %A %I:%M:%S %p')
```

```python
# Display basic summary statistics
summary_stats = terminal_df.describe()

# Display the number of unique values in each column
unique_counts = terminal_df.nunique()

# Display the data types of each column
data_types = terminal_df.dtypes

# Display missing values count in each column
missing_values = terminal_df.isnull().sum()

# Display summary statistics
print("Summary Statistics:")
print(summary_stats)

# Display unique value counts
print("\nUnique Value Counts:")
print(unique_counts)

# Display data types
print("\nData Types:")
print(data_types)

# Display missing values
print("\nMissing Values:")
print(missing_values)
```

```python
terminal_df=terminal_df.drop(["Extract_Date"], axis=1)
```

```python
terminal_df.Report_Period= pd.to_datetime(terminal_df.Report_Period)
terminal_df["Arrival_Departure"]=terminal_df["Arrival_Departure"].astype("category")
terminal_df["Domestic_International"]=terminal_df["Domestic_International"].astype("category")
terminal_df["Terminal"]=terminal_df["Terminal"].astype("category")
```

```python
terminal_df.dtypes
```

```python
data1=terminal_df.set_index("Report_Period")
```

```python
data1.head()
```

EDA for the terminal_df

```python
# Explore unique values in each categorical column
for column in data1.select_dtypes(include='object').columns:
    print(f"\nUnique values in {column}:")
    print(data1[column].value_counts())

# Plot histograms for numerical columns
data1['Passenger_Count'].plot(kind='hist', bins=20, edgecolor='black', figsize=(8, 5))
plt.title('Distribution of Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Frequency')
plt.show()

# Plot count plots for categorical columns
categorical_columns = data1.select_dtypes(include='object').columns
for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=data1, x=column)
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Pair plot for numerical columns (scatter plots and histograms)
sns.pairplot(data1, hue='Domestic_International', diag_kind='hist', height=2, aspect=2)
plt.suptitle('Pair Plot of Numerical Columns', y=2)
plt.show()

# Box plot for Passenger_Count based on Arrival_Departure
plt.figure(figsize=(10, 6))
sns.boxplot(data=data1, x='Arrival_Departure', y='Passenger_Count')
plt.title('Box Plot of Passenger Count by Arrival/Departure')
plt.xlabel('Arrival/Departure')
plt.ylabel('Passenger Count')
plt.show()
```

Plot the passanger count for each terminal

```python
#Bar plot of passanger_Count by Terminal
plt.figure(figsize=(12, 6))
sns.barplot(data=data1, x='Terminal', y='Passenger_Count')
plt.title('Passenger Count by Terminal')
plt.xlabel('Terminal')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45)  
plt.show()
```

Arrival Departure passenger count by terminal

```python

plt.figure(figsize=(12,6))
sns.barplot(data=data1, x='Terminal', y='Passenger_Count', hue='Arrival_Departure', palette='viridis')
plt.title("Passenger Count by Terminal and Arrival/Departure")
plt.xlabel('Terminal')
plt.ylabel('Passenger_Count')
plt.legend(title='Arrival/Departure', loc='upper right')
plt.xticks(rotation=45)
plt.show
```

Domestic/International passenger count by terminal

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data1, x='Terminal', y='Passenger_Count', hue='Domestic_International', palette='muted')
plt.title('Passanger Count by Terminal and Domestic/International')
plt.xlabel('Terminal')
plt.ylabel('Passenger Count')
plt.legend(title='Domestic/International', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```
